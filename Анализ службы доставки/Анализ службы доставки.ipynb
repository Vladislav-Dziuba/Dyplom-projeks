```python
import pandas as pd
from datetime import datetime
import geopy
from geopy.distance import geodesic
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = 9,7
```

```python
df = pd.read_csv("1. orders coordinates.csv")
```

- id – номер заказа
- delivery_method – тип доставки (0 — мобильный курьер; 1 — пеший курьер)
- publication_time – время заказа
- provider_lat – широта заведения
- provider_lon – долгота заведения
- recipient_lat – широта точки доставки заказа
- recipient_lon – долгота точки доставки заказа


```python
### переведем строчный формат в дататайм и вытащим из него только время
df.publication_time = df.publication_time.apply(lambda x: datetime.time(datetime.strptime(x,'%Y-%m-%d %X')))  

### с помощью библиотеки geopy определим расстояние между точками 
df['distance'] = df.apply(lambda x: geodesic ((x.provider_lat,x.provider_lon), (x.recipient_lat,x.recipient_lon)). m, axis=1)
df
```

```python
### создадим для удобства верхнюю и нижнюю границу времени в дататайм и по ним отфильтруем датафрейм

low_limit =  datetime.time(datetime.strptime('16:00:00','%X'))
hight_limit =  datetime.time(datetime.strptime('20:00:00','%X'))

### в итоге получаем датафрейм с нужными нами id, столбцы publication_time и distance оставим для контроля 
df[((df["publication_time"]>low_limit) & \
                            (df["publication_time"]<hight_limit)) & (df.distance<=1500)][['id','publication_time','distance']]
```
